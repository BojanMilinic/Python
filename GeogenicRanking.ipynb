# Order of Data
figR or N indicates regional or national
First five plots are human alteration variables (1-5)
Next 3 plots are system condition/response variables (6-8)
Final 4 plots are historic disparities variables (9-12)

```python
#REGIONAL
figR, axs = plt.subplots(6, 2, figsize= figsize)

for ax, q, r, s in zip(axs.flat, data, datar, titles):
    #CREATING THE FIGURE
    ax.plot(df.region, q, color = c1, marker = m1, ls= ' ', ms=ds1, zorder=2.5, label ="lowest regional rank (>3)", mec=oc, mew=ot) 
    ax.plot(dfr.region, r, color = c3, marker = m3, ls= ' ', ms=ds3, zorder=2.5, label ="highest regional rank (1-3)", mec=oc, mew=ot)  
    #EDITING THE TICK MARKS
    ax.set_xticks(np.arange(1,19,1))
    ax.tick_params(axis='both', which='major', labelsize=font)
    ax.tick_params(axis='both', which='minor', labelsize=font)
    #AXIS LABELS
    ax.set_ylabel("Variable Percentile Rank", fontsize = font)
    ax.set_title(s, fontsize = font)
    #VIOLIN PLOT
    vdataset = []
    for x in np.arange (1,19,1): #go through each of the 18 regions one-by-one, cut out the needed data, and append it to a violin dataset
        dfv = pd.concat([q.loc[df.region == x], r.loc[dfr.region == x]])
        vdataset.append(dfv)        
    vplot = ax.violinplot(vdataset)
    # edit vplot body color
    for pc in vplot['bodies']:
        pc.set_facecolor('gray')
        pc.set_edgecolor('gray')
    # edit vplot lines/statistic marks color
    for partname in ('cbars','cmins','cmaxes'):
        vp = vplot[partname]
        vp.set_edgecolor('white')
        vp.set_linewidth(0)
    ax.grid(axis = 'y')
        
# CHARACTERISTICS UNIQUE TO INDIVIDUAL GRAPHS
# setting up legend to have order of parts changed
handles, labels = axs[5,1].get_legend_handles_labels() 
# new legend order
order = [1,0] 
#ax[xvalues[-1]].legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize = font, bbox_to_anchor=(1.00,-1.4)) #placing bottom right under the x axis
#axs[0,1].legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize = font, bbox_to_anchor=(1.01, 1.01)) #placing top right to the right of the yaxis
axs[0,1].legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize = font, bbox_to_anchor=(0.45, 1.11)) #placing to the top right above the x axis/next to title

axs[5,0].set_xlabel("Hydrologic Region", fontsize = font) #only needed in bottom of graph
axs[5,1].set_xlabel("Hydrologic Region", fontsize = font)
axs[5,0].set_xticklabels(ticklabels, rotation = 60, horizontalalignment='right')
axs[5,1].set_xticklabels(ticklabels, rotation = 60, horizontalalignment='right')

plt.tight_layout()
```

```python
#NATIONAL but with only two categories CONCAT
figN, axs = plt.subplots(6, 2, figsize= figsize)
for ax, q, r, s, t in zip(axs.flat, datanC, datanB, titles, datanA):
    #CREATING THE FIGURE
    ax.plot(pd.concat([dfnC.region,dfnB.region]), pd.concat([q,r]), color = c1, marker = m1, ls= ' ', ms=ds1, zorder=2.5, label ="lowest national rank (55-163)", mec=oc, mew=ot) 
    ax.plot(dfnA.region, t, color = c3, marker = m3, ls= ' ', ms=ds3, zorder=2.5, label ="highest national rank (1-54)", mec=oc, mew=ot) 
    #EDITING THE TICK MARKS
    ax.set_xticks(np.arange(1,19,1))
    ax.tick_params(axis='both', which='major', labelsize=font)
    ax.tick_params(axis='both', which='minor', labelsize=font)
    #AXIS LABELS
    ax.set_ylabel("Variable Percentile Rank", fontsize = font)
    ax.set_title(s, fontsize = font)
    #VIOLIN PLOT
    vdataset = []
    for x in np.arange (1,19,1): #go through each of the 18 regions one-by-one, cut out the needed data, and append it to a violin dataset
        dfv = pd.concat([q.loc[dfnC.region == x],r.loc[dfnB.region == x],t.loc[dfnA.region == x]])
        vdataset.append(dfv)        
    vplot = ax.violinplot(vdataset)
    # edit vplot body color
    for pc in vplot['bodies']:
        pc.set_facecolor('gray')
        pc.set_edgecolor('gray')
    # edit vplot lines/statistic marks color
    for partname in ('cbars','cmins','cmaxes'):
        vp = vplot[partname]
        vp.set_edgecolor('white')
        vp.set_linewidth(0)
    #ax.grid(True, linewidth = 1, color = 'gray', linestyle='--')#color='r', linestyle='-', linewidth=2)
    ax.grid(axis = 'y')

#CHARACTERISTICS UNIQUE TO INDIVIDUAL GRAPHS
handles, labels = axs[5,1].get_legend_handles_labels() #setting up legend to have order of parts changed
order = [1,0] #new legend order
#ax[xvalues[-1]].legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize = font, bbox_to_anchor=(1.00,-1.4))
axs[0,1].legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize = font, bbox_to_anchor=(0.40,1.11))

axs[5,0].set_xlabel("Hydrologic Region", fontsize = font) #only needed in bottom of graph
axs[5,1].set_xlabel("Hydrologic Region", fontsize = font)
axs[5,0].set_xticklabels(ticklabels, rotation = 60, horizontalalignment='right')
axs[5,1].set_xticklabels(ticklabels, rotation = 60, horizontalalignment='right')

plt.tight_layout()
```
